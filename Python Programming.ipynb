```python
import random 
import numpy as np
import pandas as pd
import re
```

```python
# Code untuk generate randomly data-data random

def generate_username():
    #polanya 9 digit alphabetic + _ + 3 numeric
    list_vocal = ["a","i","u","e","o"]
    list_consonant = list(set(map(chr, range(ord('a'), ord('z')+1))).difference(set(list_vocal)))
    list_number = [str(x) for x in range(10)]
    
    acak_pertama = [random.choice(list_consonant) if random.randint(1,3) == 1 else random.choice(list_vocal) for x in range(9)]
    acak_kedua = [random.choice(list_number) for x in range(3)]
    
    return("".join(acak_pertama)+"_"+"".join(acak_kedua))


def generate_email():
    #polanya 15 digit alphabetic + _ + 3 numeric + @ + gmail/hotmail/yahoo.com
    list_vocal = ["a","i","u","e","o"]
    list_consonant = list(set(map(chr, range(ord('a'), ord('z')+1))).difference(set(list_vocal)))
    list_number = [str(x) for x in range(10)]
    list_email = ["gmail.com","hotmail.com","yahoo.com","outlook.com"]
    
    acak_pertama = [random.choice(list_vocal) if random.randint(1,3) == 1 else random.choice(list_consonant) for x in range(15)]
    acak_kedua = [random.choice(list_number) for x in range(3)]
    acak_ketiga = random.choice(list_email)
    
    return("".join(acak_pertama)+"_"+"".join(acak_kedua) + "@" + acak_ketiga)

def generate_phone():
    #08 56/98/12/... + 8 digit random
    list_provider = ["12","18","19","59","77","78",'38','31','32','33','95','96','97','98','99','56','57','58','55','12','22','13']
    list_number = [str(x) for x in range(10)]
    
    acak_8_digit = [random.choice(list_number) for x in range(8)]
    
    return("08"+random.choice(list_provider)+"".join(acak_8_digit))

kota_jabar = """Kabupaten Bandung
                Kabupaten Bandung Barat
                Kabupaten Bekasi
                Kabupaten Bogor
                Kabupaten Ciamis
                Kabupaten Cianjur
                Kabupaten Cirebon
                Kabupaten Garut
                Kabupaten Indramayu
                Kabupaten Karawang
                Kabupaten Kuningan
                Kabupaten Majalengka
                Kabupaten Pangandaran
                Kabupaten Purwakarta
                Kabupaten Subang
                Kabupaten Sukabumi
                Kabupaten Sumedang
                Kabupaten Tasikmalaya
                Kota Bandung
                Kota Banjar
                Kota Bekasi
                Kota Bogor
                Kota Cimahi
                Kota Cirebon
                Kota Depok
                Kota Sukabumi
                Kota Tasikmalaya""".split("\n")
```

```python
# Set Seed
np.random.seed(29102020) #29-10-2020

# Mau ada berapa user?
n = 100

# Do not touch
data_homework = pd.DataFrame({"buyer_id":list(range(1,n+1,1)),
                          "username": [generate_username() for x in range(n)],
                          "email" : [generate_email() for x in range(n)],
                          "phone": [generate_phone() for x in range(n)],
                          "city": [random.choice(kota_jabar) for x in range(n)],
                          "trx": [random.randint(1,100) for x in range(n)],
                          "ltv": [random.randint(1000,10000000) for x in range(n)],
                          "total_visit": [random.randint(1,100) for x in range(n)]
                         })
data_homework
```

```python
def is_prime(x):
    if x < 4:
        return True
    if all([(x > 2), (x % 2 == 0)]):
        return False
    else:
        return np.array([*map(lambda y: ((x % y) == 0).sum(), np.arange(1, x + 1))]).sum() == 2
```

```python
data_homework['voucher_discount']=data_homework["trx"].map(lambda x: 'layak' if is_prime(x)==True else 'tidak layak')
data_homework
```
